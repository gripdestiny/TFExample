```python
#-*- encoding:utf-8 -*-
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize']=(15.,8.0)
plt.rcParams['image.interpolation']='nearest'
plt.rcParams['image.cmap']='gray'
```

```python
#摘自CS231n课程作业，参考cifar10官网做出小修改
def load_CIFAR_batch(filename):
  """ load single batch of cifar """
  with open(filename, 'rb') as f:
    datadict = pickle.load(f, encoding='bytes')#添加encoding='bytes'
    #print(datadict.keys())
    X = datadict[b'data']#添加b
    Y = datadict[b'labels']#添加b
    X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("float")
    Y = np.array(Y)
    return X, Y

def load_CIFAR10(ROOT):
  """ load all of cifar """
  xs = []
  ys = []
  for b in range(1,6):
    f = os.path.join(ROOT, 'data_batch_{}'.format(str(b)))
    #print(f)
    X, Y = load_CIFAR_batch(f)
    xs.append(X)
    ys.append(Y)    
  Xtr = np.concatenate(xs)
  Ytr = np.concatenate(ys)
  del X, Y
  Xte, Yte = load_CIFAR_batch(os.path.join(ROOT, 'test_batch'))
  return Xtr, Ytr, Xte, Yte
```

```python
cifar10_dir = 'datasets/cifar-10-batches-py'
X_train,y_train,X_test,y_test = load_CIFAR10(cifar10_dir)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#将label转成one-hot vector
def one_hot(input_labes):
    out_labels=[]
    for i in range(len(input_labes)):
        label = 10*[0]
        label[input_labes[i]]=1
        out_labels.append(label)
    return np.array(out_labels)
#y_train_onehot = one_hot(y_train)
#y_test_onehot = one_hot(y_test)
```

```python
#可视化数据
classes = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
num_classes = len(classes)
samples_per_class = 5
for i,cls in enumerate(classes):
    idxs = np.flatnonzero(y_train==i)#返回label为i的位置
    idxs = np.random.choice(idxs,samples_per_class,replace=False)#随机图像
    for j, idx in enumerate(idxs):
        plt_idx = j * num_classes + i + 1
        plt.subplot(samples_per_class,num_classes,plt_idx)
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if j==0:
            plt.title(cls)
plt.show()
```
